```python
x=np.array(dataset['YearsExperience']).reshape(-1,1)
y=np.array(dataset['Salary']).reshape(-1,1)
l=len(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,explained_variance_score
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=6)
lx=len(x_train)
```

# Lasso regression

```python
def lasso(alpha):
    lasso_reg=Lasso(alpha=alpha)
    lasso_reg.fit(x_train,y_train)
    y_pred=lasso_reg.predict(x_test)
    error=pd.DataFrame([[mean_squared_error(y_test,y_pred),
                    mean_absolute_error(y_test,y_pred),
                    explained_variance_score(y_test,y_pred)]],
                   columns=["mean_squared_error",
                                 "mean_absolute_error","explained_variance_score"])
    print(error)
    plt.plot(x_test,y_pred,marker='o',
         color='blue',markerfacecolor='red',
         markersize=10,linestyle='dashed')
    plt.scatter(x,y,marker='o',color='red')
    plt.xlabel("YearsExperience")
    plt.ylabel("salary")
    plt.title(f"alpha=={alpha}")
    plt.show()
```

```python
lasso(0.1)
```

```python
lasso(0.6)
```

```python
lasso(10000000000000000)
```

# Elastic net

```python
def elasticnet(alpha,r):
    elastic=ElasticNet(alpha=alpha,l1_ratio=r)
    elastic.fit(x_train,y_train)
    y_pred=elastic.predict(x_test)
    error=pd.DataFrame([[mean_squared_error(y_test,y_pred),
                    mean_absolute_error(y_test,y_pred),
                    explained_variance_score(y_test,y_pred)]],
                   columns=["mean_squared_error",
                                 "mean_absolute_error","explained_variance_score"])
    print(error)
    plt.plot(x_test,y_pred,marker='o',
         color='blue',markerfacecolor='red',
         markersize=10,linestyle='dashed')
    plt.scatter(x,y,marker='o',color='red')
    plt.xlabel("YearsExperience")
    plt.ylabel("salary")
    plt.title(f"alpha=={alpha},r=={r}")
    plt.show()
```

```python
elasticnet(0.5,10)
```

```python
elasticnet(0.5,5)
```

```python
elasticnet(0.5,2)
```

```python
elasticnet(0.5,1000)
```
