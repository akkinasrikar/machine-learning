```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
dataset["label"]=encoder.fit_transform(dataset["label"])
```

```python
x=dataset.drop(["label"],axis=1)
y=dataset["label"]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=12)
```

```python
x_train.shape[1]
```

```python
class Logisticregression:
    def __init__(self,learning_rate,epochs):
        self.learning_rate=learning_rate
        self.epochs=epochs
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
    def fit(self,x_data,y_data):
        x_new_data=np.c_[np.ones((x_data.shape[0],1)),x_data]
        self.theta=np.zeros(x_new_data.shape[1])
        for i in range(self.epochs):
            data=np.dot(x_new_data,self.theta)
            cost_function=np.dot(x_new_data.T,self.sigmoid(data)-y_data)
            gradients=cost_function/len(x_data)
            self.theta=self.theta-self.learning_rate*gradients
    def predict(self,x_test_data):
        x_test_new=np.c_[np.ones((x_test_data.shape[0],1)),x_test_data]
        self.results=self.sigmoid(np.dot(x_test_new,self.theta))
        self.y_pred=[0 if i<0.5 else 1 for i in self.results]
        return self.y_pred
    def plot_prob(self):
        pl=sorted(self.results,reverse=False)
        plt.figure(figsize=(10,6))
        plt.plot(pl,marker='o',
         color='blue',markerfacecolor='red',
         markersize=5)
        plt.ylabel("probability")
        plt.title("Logistic Regression")
        plt.show()
    def plot_cost(self):
        class1=[]
        class0=[]
        for i in range(len(self.y_pred)):
            if self.y_pred[i]==1:
                class1.append(-np.log(self.results[i]))
            else:
                class0.append(1-np.log(self.results[i]))
        ax=plt.figure(figsize=(10,6))
        ax.add_subplot(121)
        plt.plot(sorted(class0,reverse=False),
                marker='o',color='blue',markerfacecolor='red',
                markersize=8)
        plt.title("class0")
        ax.add_subplot(122)
        plt.plot(sorted(class1,reverse=True),marker='o',
                 color='blue',markerfacecolor='red',
                 markersize=8)
        plt.title("class1")
        plt.show()                              
```

```python
model=Logisticregression(0.1,10000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(accuracy_score(y_test,y_pred)*100)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,yticklabels=False)
```
